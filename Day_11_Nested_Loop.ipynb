```python
#Nested loop_Loop inside loop
```

```python
rows = 4
#outer loop
for x in range(1, rows+1):
    #inner loop
    for y in range(1, 1+x):
        print('#', end=' ')
    print('')
```

```python
num = 1,2,3,4
#outer loop
for ele in num:
    #inner loop
    index = 0
    while index < len(num):
        print(num[index], end = ' ')
        index = index + 1
    print('\n')
```
