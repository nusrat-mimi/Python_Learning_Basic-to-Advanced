# 1.Take values of Length and Breadth of a rectangle from user input() and check if it is square or not.

```python
#Take length and breadth as input
```

```python
length = int(input('Enter the length: '))
breadth = int(input('Enter the breadth: '))
```

```python
#check if it is square or not
```

```python
if length == breadth:
    print('The rectangle is square.')
else:
    print('The rectangle is not square.')
```

# *3. A student will not be allowed to seat in exam if his/her attendance is less than 75%*

```python
#input attendence percentage
```

```python
held = int(input('Number of classes held: '))
attendance = int(input('Enter attendance percentage '))
attended_ratio = (attendance/held)*100
```

```python
attended_ratio
```

```python
if attended_ratio >=75 :
    print('Student is allowed to seat in the exam')
else:
    print('Student is not allowed to seat in the exam')
```

# *4. A school has following rules for grading system:*
Below 25 - fail, 25 to 44 - E, 45 to 49 - D, 50 to 59 - C, 60 to 79 - B, 80 to 89 - A, Above 90 - A+
now ask user to enter marks and print the corresponding grade

```python
mark = float(input('Enter mark: '))
```

```python
if mark < 25:
    grade ='Fail'
elif 25 <= mark <=44:
    grade ='E'
elif 45 <= mark <= 49:
    grade ='D'
elif 50 <= mark <= 59:
    grade ='C'
elif 60 <= mark <= 79:
    grade ='B'
elif 70 <= mark <= 89:
    grade = 'A'
else:
    grade ='A+'

print('My grade is: ',grade)
```
