```python
name = ['nusrat', 'jahan', 'mimi']
varsity = ['cu','du','freie']
```

```python
# ('nusrat','cu'), ('jahan','du',), .............
zipped = zip(name, varsity)
print(zipped)
```

```python
zipped = list(zip(name, varsity))
print(zipped)
```

```python
zipped = tuple(zip(name, varsity))
print(zipped)
```

```python
zipped = dict(zip(name, varsity))
print(zipped)
```

```python
zipped.keys()
```

```python
zipped.values()
```

```python
zipped = set(zip(name, varsity))
print(zipped)
```

# unzip

```python
zipped = list(zip(name, varsity))
print(zipped)
```

```python
name1, varsity1 = zip(*zipped)
```
