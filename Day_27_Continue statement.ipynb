```python
sen = 'Ich liebe Bangladesh und Germany'

for i in sen.split():
    if i == 'und':
        continue
    print(i)
```

```python
#using while loop and continue statement
```

```python
num = (1, 2, 3, 4, 5, 6, 10, 14, 15)

i = 0

while i < len(num):
    if num[i] % 2 == 1:
        i += 1
        continue
    print(num[i])
    i += 1
```

```python
sen = 'Ich liebe Bangladesh und Germany'

words = sen.split()
i = 0

while i < len(words):
    if words[i] == 'und':
        i += 1
        continue

    print(words[i])
    i += 1
```
